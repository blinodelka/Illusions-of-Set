```python
import numpy as np
import keras
from keras.utils import to_categorical
```

How to decode information received from a noisy channel?

```python
adaptation_data = np.concatenate((np.random.normal(3, 0.3, 20), np.random.normal(2,0.3, 20), np.random.normal(1,0.3,20),
                                 np.random.normal(0, 0.3, 20), np.random.normal(-1, 0.3, 20),np.random.normal(-2, 0.3, 20),
                                 np.random.normal(-3, 0.3, 20)))

true_answers = np.concatenate((np.repeat(3, 20), np.repeat(2, 20), np.repeat(1, 20), np.repeat(0, 20),
                              np.repeat(-1, 20), np.repeat(-2, 20), np.repeat(-3, 20)))
```

```python
# as first layer in a sequential model:
model = keras.Sequential()
model.add(keras.layers.Dense(7, input_shape=(1,), kernel_regularizer=keras.regularizers.l2(0.5), 
                             activity_regularizer = keras.regularizers.l2(0.01)))
model.add(keras.layers.Dense(1, kernel_regularizer=keras.regularizers.l2(0.2)))
model.add(keras.layers.Dense(1, kernel_regularizer=keras.regularizers.l2(0.2)))
# now the model will take as input arrays of shape (*, 16)
# and output arrays of shape (*, 32)
```

```python
model.compile(loss='mean_squared_error', optimizer= 'adam')
```

```python
#sgd = keras.optimizers.SGD(lr=0.005)

model.fit(adaptation_data, true_answers, epochs=600, batch_size=10)
```

```python
predictions = model.predict(adaptation_data)
```

classifier

```python
adaptation_data = np.concatenate((np.random.normal(3, 0.3, 20), np.random.normal(2,0.3, 20), np.random.normal(1,0.3,20),
                                 np.random.normal(0, 0.3, 20), np.random.normal(-1, 0.3, 20),np.random.normal(-2, 0.3, 20),
                                 np.random.normal(-3, 0.3, 20)))

true_classes = np.concatenate((np.repeat(0, 20), np.repeat(1, 20), np.repeat(2, 20), np.repeat(3, 20),
                              np.repeat(4, 20), np.repeat(5, 20), np.repeat(6, 20)))

true_classes_oh = to_categorical(true_classes)
print(true_classes_oh)
```

```python
# invert encoding
inverted = np.argmax(true_classes_oh, axis = 1)
print(inverted)
```

```python
classifier = keras.Sequential()
classifier.add(keras.layers.Dense(3, input_shape=(1,)))
classifier.add(keras.layers.Dense(7, init='uniform'))
classifier.add(keras.layers.Activation('sigmoid'))
               
```

```python
sgd = keras.optimizers.SGD(lr=0.05, decay=1e-6, momentum=0.9, nesterov=True)
classifier.compile(loss='binary_crossentropy', optimizer=sgd)
```

```python
classifier.fit(adaptation_data, true_classes_oh, epochs=1000, batch_size=50)
```

```python
np.argmax(classifier.predict(adaptation_data), axis = 1)
```

```python
adaptation_trial = np.random.normal(-3, 0.3, 2)

adaptation_true_answers = np.repeat(6, 2)
adaptation_true_answers_oh = to_categorical(adaptation_true_answers)

test_trial = np.array([0])
```

```python
np.argmax(classifier.predict(test_trial), axis = 1)
```

```python
classifier.fit(adaptation_trial, adaptation_true_answers_oh, epochs=5, batch_size=2)
```

```python
np.argmax(classifier.predict(adaptation_trial), axis = 1)
```
